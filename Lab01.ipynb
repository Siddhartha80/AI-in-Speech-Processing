```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
file_path = 'Lab01.mp3'
y, sr = librosa.load(file_path)

# Plot the speech signal
librosa.display.waveshow(y, sr=sr)
plt.title('Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

threshold = 0.0002
silence_mask = np.abs(y) < threshold
silence_duration = np.sum(silence_mask) / sr

print(f'Silence duration between words: {silence_duration:.2f} seconds')
```

### Playing Small Segment of Signal

```python

import IPython.display as ipd
file_path = 'Lab01.mp3'
y, sr = librosa.load(file_path)

start_time = 1.5
end_time = 3    
segment = y[int(start_time*sr):int(end_time*sr)]

# Play the segment
ipd.Audio(segment, rate=sr)
```

```python
librosa.display.waveshow(segment,sr=sr)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
y, sr = librosa.load('Lab01.mp3')

# Segment the speech signal abruptly (for example, every 5 seconds)
segment_length = 1 * sr
segments = [y[i:i+segment_length] for i in range(0, len(y), segment_length)]

# Display each segment separately
for i, segment in enumerate(segments):
    plt.figure()
    librosa.display.waveshow(segment)
    plt.title(f'Speech Signal - Segment {i+1}')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.show()
```
